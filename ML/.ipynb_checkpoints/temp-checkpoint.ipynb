```python
numerical_cols=df.select_dtypes(include=[np.number]) #will contain columns with numerical data types.
categorical_cols=df.select_dtypes(include=['object']) #will contain columns with object data type (assuming they are categorical).
```

```python
# Fill missing values with the mean of the column
df['Credit_History'].fillna(df['Credit_History'].mean(),inplace=True)
```

```python
# Drop rows with any missing values
df.dropna()
```

```python
#this drops all the duplicate rows
df=df.drop_duplicates()
```

```python
#Data Manipulation:
#You can perform various operations on the data, such as filtering, sorting, grouping, and aggregating.
```

```python
#Feature Selection
#Identify relevant features for training the model based on the problem statement and dataset.
X = df[['CoapplicantIncome', 'LoanAmount','Loan_Amount_Term','Credit_History']]
Y = df['Loan_Status']
```

```python
#Split the dataset into training and testing sets to evaluate the model's performance.
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.75,random_state=10)
```

```python
#Building Decision Tree
#Create and train a Decision Tree model using a library like scikit-learn.
from sklearn.tree import DecisionTreeClassifier
```

```python
#Training the model
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
#Make predictions on the testing set
Y_pred=model.predict(X_test)
```

```python
#Model Evaluation
from sklearn import metrics, preprocessing
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
precision = precision_score(Y_test,Y_pred)
recall = recall_score(Y_test,Y_pred)
f1_score = f1_score(Y_test,Y_pred)
```

```python
print(f"Accuracy : {accuracy}")
print(f"Precision : {precision}")
print(f"Recall : {recall}")
print(f"F1 Score : {f1_score}")
```

```python
cm = confusion_matrix(Y_test,Y_pred)
print("Confusion matrix:-")
print(cm)
```
